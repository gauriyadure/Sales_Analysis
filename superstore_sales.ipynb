### Q1. What is the overall sales trend? 

```python
d['order_date'].min()
```

```python
d['order_date'].max()
```

```python
d['month_year']=d['order_date'].apply([lambda x: x.strftime('%Y-%m')])
d['month_year']
```

```python
d.groupby('month_year').sum()
```

```python
d_trend=d.groupby('month_year').sum()['sales'].reset_index()
d_trend
```

```python
plt.figure(figsize=(15,7))
plt.bar(d_trend['month_year'],d_trend['sales'])
plt.xlabel('Month_year')
plt.ylabel('Sales')
plt.xticks(rotation='vertical')
plt.show()
```
